### read stored data from DataPlatform

```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

```python
dir_raw_id = 106
dir_process_id = 107
```

```python
datanode = DataNode()
```

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='skooldio.json'))
meta
```

```python
data = json.load(fp)
data
```

```python
df2 = pd.json_normalize(data)
```

```python
df = pd.DataFrame(df2)
```

### write DataNode to DataPlatform

```python
datanode.write(df=df2, directory=dir_process_id, name="skooldio", profiling=True, lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='skooldio.json')])
```
